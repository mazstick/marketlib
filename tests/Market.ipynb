```python
df = pd.read_csv("E:/Work/data/XAUUSD/XAUUSD-m1.csv")
x = df[:1000].copy()
x
```

```python
a = 0.00028
n = 2
df = data.Market(x, "XAUUSD", "m1").to_dataframe()
bb = indicators.BollingerBands(df, 18, 2)
bb_advance = bb.calculate()

bb_advance["upper"] = bb_advance["upper"] - df["open"]
bb_advance["middle"] = bb_advance["middle"] - df["open"]
bb_advance["lower"] = bb_advance["lower"] - df["open"]

diff = bb()["upper"] - bb()["lower"]
diff_per_price = (diff / df["open"]) * 100

h_diff = df["open"] * (1.0 + a * n)
l_diff = df["open"] * (1.0 - a * n)

h_diff = h_diff - df["open"]
l_diff = l_diff - df["open"]

df["close"] = df["close"] - df["open"]
df["high"] = df["high"] - df["open"]
df["low"] = df["low"] - df["open"]
df["open"] = df["open"] - df["open"]

ap1 = mpf.make_addplot(bb_advance, panel = 0, color='black', secondary_y=False)
ap2 = mpf.make_addplot(diff_per_price, panel=1, color = 'r', ylabel="diff/price * 100")
ap3 = mpf.make_addplot(h_diff, panel = 0, label=f"open * 1 + {a} * {n}", color='b',secondary_y=False)
ap4 = mpf.make_addplot(l_diff, panel = 0, label=f"open * 1 - {a} * {n}", color='r',secondary_y=False)
mpf.plot(df,style="yahoo",addplot=[ap2, ap3, ap4], figratio=(30,10), figscale=1.5, type='candle')
```
