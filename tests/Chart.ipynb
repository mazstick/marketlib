```python
from marketlib import Chart
from marketlib.data import Market
from marketlib import indicators
import pandas as pd
```

```python
df = pd.read_csv("E:/Work/data/XAUUSD/XAUUSD-m5.csv")
df = df[:400]
df
```

```python
xauusd = Market(df, "XAUUSD", "5m")
sma = indicators.SMA(df, 24)
bb = indicators.BollingerBands(df, 30, 2)
atr = indicators.ATR(df, 14)
atr_sma = indicators.ATR(df, 14, 'sma')
atr_ema = indicators.ATR(df, 14, fast_wilder=True)
atr_ema.layer.set_layer(color='g',label="atr ema")
atr_sma.layer.set_layer(color='r', label="atr sma")
```

```python
chart = Chart()
atr.layer.ylabel = "ATR"
chart.add_market(xauusd)
chart.add_indicator(atr_sma)
chart.add_indicator(atr_ema)
chart.plot()
```

```python
chart = Chart()
atr.layer.ylabel = "ATR"
chart.add_market(xauusd)
chart.add_indicator(atr_sma)
chart.add_indicator(sma)
sma.layer.panel = 2
bb.layer.panel = 0
chart.add_indicator(bb)
chart.add_fiilbetween(y1=bb()["upper"], y2=bb()["middle"], where=(df.index > "2025-05-23 02:55:00"), color='r', interpolate=False)
chart.add_fiilbetween(y1=bb()["lower"], y2=0, where=(xauusd.to_dataframe().index < "2025-05-23 02:55:00") & (xauusd.to_dataframe().index > "2025-05-22 09:55:00"), color='b', interpolate=False)
chart.add_fiilbetween(y1=bb()["lower"], y2=bb()["upper"], where=(xauusd.to_dataframe().index < "2025-05-22 09:55:00"), color='g', interpolate=False, panel=0)
chart.add_fiilbetween(y1=3250, y2=sma()["sma_24"].values, color='g', panel=2)
chart.add_fiilbetween(y1=xauusd.get_lows(), color='black', alpha=.2)
chart.add_fiilbetween(y1=atr_sma.calculate(), y2=4,where= atr_sma() < 4, color='yellow', alpha=.2, panel=1)
chart.add_fiilbetween(y1=atr_sma.calculate(), y2=4,where= atr_sma() > 4, color='b', alpha=.2, panel=1)
chart.add_line(4, "hline", linestyle='--', color='r', panel=1, linewidth=1,)
chart.plot()
```
