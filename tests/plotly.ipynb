```python
import plotly.graph_objects as go
from plotly.subplots import make_subplots
import pandas as pd
from marketlib import Market

df = pd.read_csv("BTCUSDT15m.csv")
btc = Market(df, "BTCUSDT", "15m")
btc.to_dataframe()
```

```python
fig = make_subplots(rows=1, cols=1, shared_xaxes=True, vertical_spacing=0.5)
fig.add_trace(
    go.Candlestick(
        x=btc.to_dataframe().index,
        close=btc.get_closes(),
        open=btc.get_opens(),
        high=btc.get_highs(),
        low=btc.get_lows(),
        increasing=go.candlestick.Increasing(fillcolor="rgb(0,200,0)"),
        decreasing=go.candlestick.Decreasing(fillcolor="rgb(200,0,0)"),
    )
)

fig.update_layout()

fig.show()
```

```python
from marketlib.plotly import Plotly
from marketlib.indicators import *
sma = SMA(btc.to_dataframe(), [14, 60])
plotly = Plotly()
plotly.add_chart(btc.to_dataframe(), btc.symbol, btc.timeframe)
plotly.add_indicators(sma)
plotly.plot()
```
