```python
from scipy.signal import find_peaks

macd_peak, _ = find_peaks(macd().iloc[:, 0].abs(), distance=15)
price_peak_down, _ = find_peaks(-btcusdt.to_dataframe()["low"],prominence=60, distance=7)
price_peak_up, _ = find_peaks(btcusdt.to_dataframe()["high"].tolist(),prominence=60, distance=7)
macd_peak_up = []
macd_peak_down = []
for x in macd_peak.tolist():
    if macd_df.iloc[x]["macd"] < 0:
        macd_peak_down.append(x)
    else:
        macd_peak_up.append(x)
```

```python
chart.delete_lines()
chart.add_line(price_peak_down.tolist(), "tline", tline_use="low")
chart.add_line(macd_peak_down, "tline", tline_use=0, panel=1)
chart.plot(start=300,end=500)
```

```python
def map_extrema(price_extrema: list[int], macd_extrema: list[int], max_distance: int = 6) -> dict:
    mapping = {}
    for p_idx in price_extrema:
        candidates = [m_idx for m_idx in macd_extrema if abs(m_idx - p_idx) <= max_distance]
        if not candidates:
            continue
        else:
            closest = min(candidates, key=lambda m_idx: abs(m_idx - p_idx))
            mapping[p_idx] = closest
    return mapping
```

```python
nearest_peak_up =  map_extrema(price_peak_up, macd_peak_up)
nearest_peak_up = list(nearest_peak_up.items())
nearest_peak_up

nearest_peak_down = map_extrema(price_peak_down, macd_peak_down)
nearest_peak_down = list(nearest_peak_down.items())
nearest_peak_down
```

```python
TRESHHOLD = 200
pos_rd_price_lines = []
pos_rd_macd_lines = []
pos_hd_price_lines = []
pos_hd_macd_lines = []

for i in range(1, len(nearest_peak_up)):
    p1, m1 = nearest_peak_up[i-1]
    p2, m2 = nearest_peak_up[i]
    
    if p2 - p1 > TRESHHOLD:
        continue
    
    macd1, macd2 = macd_df.iloc[m1]["macd"], macd_df.iloc[m2]["macd"]
    price1 , price2 = btcusdt.to_dataframe().iloc[p1]["high"], btcusdt.to_dataframe().iloc[p2]["high"]
    print(f"{m1} {m2} | {p1} {p2}")
    if price2 - price1 > 0 and macd2 - macd1 < 0:
        pos_rd_price_lines.append([int(p1),int(p2)])
        pos_rd_macd_lines.append([m1, m2])
        
    

        
```

```python
TRESHHOLD_pos = 60
pos_rd_price_lines = []
pos_rd_macd_lines = []
pos_hd_price_lines = []
pos_hd_macd_lines = []


for i in range(0, len(nearest_peak_up)):
    p1, m1 = nearest_peak_up[i]
    for k in range(i+1, len(nearest_peak_up)):
        p2, m2 = nearest_peak_up[k]
        if p2 - p1 > TRESHHOLD_pos:
            break
        
        macd1, macd2 = macd_df.iloc[m1]["macd"], macd_df.iloc[m2]["macd"]
        price1 , price2 = btcusdt.to_dataframe().iloc[p1]["high"], btcusdt.to_dataframe().iloc[p2]["high"]
        print(f"{m1} {m2} | {p1} {p2}")
        if price2 - price1 > 0 and macd2 - macd1 < 0:
            pos_rd_price_lines.append([int(p1),int(p2)])
            pos_rd_macd_lines.append([m1, m2])
        
        if price2 - price1 < 0 and macd2 - macd1 > 0 :
            pos_hd_price_lines.append([int(p1), int(p2)])
            pos_hd_macd_lines.append([m1, m2])
    
    
TRESHHOLD_neg = 60

neg_rd_price_lines = []
neg_rd_macd_lines = []
neg_hd_price_lines = []
neg_hd_macd_lines = []


for i in range(0, len(nearest_peak_down)):
    p1, m1 = nearest_peak_down[i]
    for k in range(i+1, len(nearest_peak_down)):
        p2, m2 = nearest_peak_down[k]
        if p2 - p1 > TRESHHOLD_neg:
            break
        
        macd1, macd2 = macd_df.iloc[m1]["macd"], macd_df.iloc[m2]["macd"]
        price1 , price2 = btcusdt.to_dataframe().iloc[p1]["low"], btcusdt.to_dataframe().iloc[p2]["low"]
        print(f"{m1} {m2} | {p1} {p2}")
        if price2 - price1 < 0 and macd2 - macd1 > 0:
            neg_rd_price_lines.append([int(p1),int(p2)])
            neg_rd_macd_lines.append([m1, m2])
        
        if price2 - price1 > 0 and macd2 - macd1 < 0 :
            neg_hd_price_lines.append([int(p1), int(p2)])
            neg_hd_macd_lines.append([m1, m2])
        
    
```

```python
from marketlib.strategy.MACDdivergence import MACDDivergence
```

```python
strategy = MACDDivergence(btcusdt.to_dataframe(), treshhold=80, macd_deep=20,interspace=0, tolerance=0.05)
chart = Chart()
chart.layer.set_layer(figsize=(20, 8))
chart.add_chart(btcusdt.to_dataframe(),btcusdt.symbol, btcusdt.timeframe)
chart.add_strategy(strategy)
chart.plot()
```
