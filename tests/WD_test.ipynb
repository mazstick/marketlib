```python
from marketlib.indicators import PATR
import pandas as pd
from marketlib import Chart
from marketlib import Market
```

```python
eth5 = Market(pd.read_csv("ETHUSDT5m.csv"), symbol="ETHUSDT", timeframe="5m")
eth1 = Market(pd.read_csv("ETHUSDT1m.csv"), symbol="ETHUSDT", timeframe="1m")
```

```python
wd = PATR(eth1.to_dataframe(),14, "close", invers=True)
```

```python
chart = Chart()
chart.add_chart(eth5.to_dataframe().loc[eth5.to_dataframe().index > eth1.to_dataframe().index[0]], eth5.symbol, eth5.timeframe)
chart.add_plot(wd(),label=f"WD {wd.period}")
chart.plot()
```

```python
import pandas as pd
import mplfinance as mpf

# داده‌ها
df1 = eth1.to_dataframe()              # 1 دقیقه
df5 = eth5.to_dataframe()              # 5 دقیقه
df5 = df5.loc[df5.index > df1.index[0]]

# محاسبه اندیکاتور روی 1 دقیقه
wd_1m = wd()   # فرض بر این که خروجی Series با ایندکس df1 است

# هماهنگ کردن کندل‌های 5 دقیقه با ایندکس 1 دقیقه
df5_on_1m = df5.reindex(df1.index)
```

```python
df5_on_1m[:40]
```

```python
window = 10
margin = .02
start = 0
end = 500
shifted = wd_1m - wd_1m.shift(1)
absolot_volatility = shifted.abs().rolling(window).sum()
real_volatility = shifted.rolling(window).sum()
chart = Chart()
chart.add_chart(df5_on_1m, "ETHUSDT", "5m")
chart.add_indicator(wd)
chart.layer.figsize =(20, 6)
chart.add_plot(absolot_volatility[start:end], panel=2, label=f"wd sum abs(vol) {window}", color="#030", ylim=(0 - margin, absolot_volatility.max() + margin))
chart.add_plot(real_volatility[start:end], panel=3, label=f"wd sum vol {window}", color="#500")
chart.add_plot(real_volatility[start:end] / absolot_volatility[start:end], panel=4, label=f"vol/abs(vol) ", color="#005")
chart.add_line(0, "hline", panel=1, linewidth=0.5, color="black")
chart.add_line(0, "hline", panel=2, linewidth=0.5, color="black")
chart.add_line(0, "hline", panel=3, linewidth=0.5, color="black")

chart.plot(end=end)
```

```python
wd_14 = PATR(eth1.to_dataframe(), 14, "close", invers=True)
wd_5 = PATR(eth1.to_dataframe(), 5, "close", invers=True)
```

```python
chart = Chart()
chart.add_chart(df5_on_1m, symbol="ETHUSDT", timeframe="5m")
chart.layer.set_layer(figsize=(30, 8),figratio=(1,2))
chart.add_indicator(wd_14)
chart.add_indicator(wd_5)
wd_5.layer.set_layer(panel=1,color="#400", label="WD_inv 5")
chart.add_line(0, "hline", linewidth=.5, color="black", panel=1)
chart.add_line(0, "hline", linewidth=.5, color="black", panel=2)
wd_14_5 = wd_5() - wd_14()
chart.add_plot(wd_14_5[:300], panel=2, color="#080", label="WD inv 5 - WD inv 14")
for i in range(0, len(df5_on_1m)):
    if i % 5 == 0:
        chart.add_line(i, "vline", panel=0, color="black", linewidth=0.5)
        chart.add_line(i, "vline", panel=1, color="black", linewidth=0.5)
        chart.add_line(i, "vline", panel=2, color="black", linewidth=0.5)
chart.plot(end=300)
```
